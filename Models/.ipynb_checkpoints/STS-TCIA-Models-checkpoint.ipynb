```python
def count_cols(DataFrame, string):
    """Conta quantas colunas tem essa string nome.

    Args:
        DataFrame (dataframe): DataFrame que será contado
        string (str): nome para pesquisar nas 

    Returns:
        int: quantidade de colunas que passaram no filtro
    """
    return len(list(col for col in DataFrame.columns if col.startswith(string)))


case_cols = count_cols(df_sts_tcia_rad, "case_")
diagnostic_cols = count_cols(df_sts_tcia_rad, "diagnostic")
original_shape_features_cols = count_cols(df_sts_tcia_rad, "original_shape_")
original_features_cols = count_cols(
    df_sts_tcia_rad, "original_") - original_shape_features_cols  # removendo colunas repetidas
log_feature_cols = count_cols(df_sts_tcia_rad, "log-")
wavelet_feature_cols = count_cols(df_sts_tcia_rad, "wavelet-")
features_cols = original_shape_features_cols + \
    original_features_cols + log_feature_cols + wavelet_feature_cols

print(f"""
Features count: {features_cols} 
    Shape: {original_shape_features_cols}
    Original: {original_features_cols}
    LoG: {log_feature_cols}
    Wavelet: {wavelet_feature_cols}
""")
```

```python
def remove_cols(DataFrame, string):
    """Remover colunas com uma string no nome

    Args:
        DataFrame (DataFrame): Dataframe que sera removido
        string (str): texto para buscar colunas para remover

    Returns:
        DataFrame: Dataframe sem colunas desejadas
    """
    list_cols = list(
        col for col in DataFrame.columns if col.startswith(string))
    return DataFrame.drop(list_cols, axis=1)


data = remove_cols(df_sts_tcia_rad, "diagnostics_")
```

```python
def tag_cols(data, prefix, excepts):
    """Adicionar tag nas colunas

    Args:
        data (DataFrame): DataFrame que sera colocado tag
        prefix (str): tag que será adicionada
        excepts (str): Colunas que não terão tags

    Returns:
        _type_: _description_
    """
    data.columns = [col if col.startswith(
        excepts) else prefix+"_" + col for col in data.columns]
    return data
```

```python
data_t1 = data.loc[data.case_ImageType == "T1"]
tag_cols(data_t1, "T1", "case_")
data_t1 = data_t1.drop(columns=["case_ID", "case_ImageType"])
```

```python
data_fs = data.loc[data.case_ImageType != 'T1']
tag_cols(data_fs, "FS", "case_")
data_fs = data_fs.drop(
    columns=['case_ID', 'case_ProviderID', 'case_ImageType'])
```

## Feature selector e modelos

Configurando `feature_selection` e `models` que serão usados

```python
from sklearn.feature_selection import mutual_info_classif, GenericUnivariateSelect, RFECV, SelectFromModel, VarianceThreshold
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import cross_validate
```

```python
X = data_merged.drop(["case_ProviderID", "case_PatientID"], axis=1)
y = np.array(df_sts_tcia_cli.LungMets)
```

```python
estimator_RFE = SVC(kernel="linear", verbose=5)
estimator_L1LSVC = LinearSVC(penalty="l2", dual=True, verbose=0)
estimator_L1LR = LogisticRegression(penalty="l2", dual=True, solver="liblinear", verbose=0)

VT_sel = VarianceThreshold(threshold=.8 * (1 - .8))
MIC_sel = GenericUnivariateSelect(score_func=mutual_info_classif, mode='k_best', param=50)
RFE_sel = RFECV(estimator_RFE, min_features_to_select=50, step=10, cv=StratifiedKFold(5), scoring='accuracy')
L1LSVC_sel = SelectFromModel(estimator_L1LSVC)
L1LR_sel = SelectFromModel(estimator_L1LR)

rf_ckf = RandomForestClassifier(n_estimators=20, bootstrap=True, min_samples_leaf=3, max_features="sqrt")
lr_clf = LogisticRegression(C=1.0, class_weight="balanced", penalty="l2")
knn_clf = KNeighborsClassifier(n_neighbors=5, p=2)
svc_clf = SVC(C=1, kernel="rbf", class_weight="balanced", probability=True)
```

```python
def get_name(model_or_selector):
    """Retira o nome do modelo ou seletor de features

    Args:
        model_or_selector (model_or_selector): String com a função toda

    Returns:
        str: Só o nome do modelo ou seletor
    """
    full_name = str(model_or_selector)
    end_name = full_name.index("(")
    name = full_name[:end_name]
    return name


def get_name_estimator(model_or_selector):
    if "estimator" in model_or_selector.get_params().keys():
        full_name = model_or_selector.get_params()["estimator"]
        return get_name(full_name)
    else:
        return None


def get_selector_and_estimator(selector):
    estimator_name = get_name_estimator(selector)
    selector_name = get_name(selector)
    if estimator_name is None:
        return selector_name
    else:
        return selector_name+"_"+estimator_name
```

```python
from time import time
selectors = [VT_sel, MIC_sel, L1LSVC_sel, L1LR_sel]
models = [rf_ckf, lr_clf, knn_clf, svc_clf]

selectors_infos = pd.DataFrame(
    columns="selector time shape".split(" "))

for selector in selectors:
    # print(str(selector), ": ")
    tic = time()
    selector.fit(X, y)
    toc = time()
    # print("\n")
    shape_X = selector.transform(X).shape
    selectors_infos.loc[len(selectors_infos)] = [get_selector_and_estimator(selector), toc-tic, shape_X]

selectors_infos
```

```python
columns_names = ["models", "selector", "f1_mean", "f1_max", "accuracy_mean", "accuracy_max",
                 "fit_time_mean", "fit_time_max", "feature_list", "selector_params", "model_params"]
results = pd.DataFrame(columns=columns_names)


scoring = ("f1", "accuracy")

all_features_name = X.columns.values.tolist()

for selector in selectors:
    features_list = selector.get_feature_names_out(all_features_name)
    X_new = selector.transform(X)
    selector_name = get_selector_and_estimator(selector)
    selector_params = selector.get_params()
    for model in models:
        model_name = get_name(model)
        model_params = model.get_params()
        result = cross_validate(model, X_new, y, scoring=scoring, verbose=0)

        test_f1 = result["test_f1"]
        f1_mean = test_f1.mean()
        f1_max = test_f1.max()

        test_accuracy = result["test_f1"]
        accuracy_mean = test_accuracy.mean()
        accuracy_max = test_accuracy.max()

        fit_time = result["fit_time"]
        fit_mean = fit_time.mean()
        fit_max = fit_time.max()

        results.loc[len(results)] = [model_name, selector_name, f1_mean, f1_max, accuracy_mean,
                                     accuracy_max, fit_mean, fit_max, features_list, selector_params, model_params]
```
